```
import json

with open('/gdrive/My Drive/PETR 370/Lab 1/PLSSTownship.GeoJSON') as geofile:
  NMTownship =json.load(geofile)
```

```
pts=[]#list of points defining boundaries of polygons

myInterest = ['26S 33E', '26S 34E', '25S 34E', '25S 33E']

for  feature in NMTownship['features']:
    if feature['geometry']['type']=='Polygon' and feature['properties']['TWNSHPLAB'] in myInterest:
        pts.extend(feature['geometry']['coordinates'][0])    
        pts.append([None, None])#mark the end of a polygon   
        
    elif feature['geometry']['type']=='MultiPolygon' and feature['properties']['TWNSHPLAB'] in myInterest:
        for polyg in feature['geometry']['coordinates']:
            pts.extend(polyg[0])
            pts.append([None, None])#end of polygon
    #else: raise ValueError("geometry type irrelevant for map")

X, Y=zip(*pts)
```
